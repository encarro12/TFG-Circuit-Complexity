```python
# Libraries used
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Here we are going to analyze the entanglement
metric = 'entanglement'
```

```python
# Import the files with the data in dataframes for its posterior analysis
dataset_df = pd.read_csv("dataset.csv")
non_dataset_df = pd.read_csv("non_dataset.csv")
```

```python
# Compare functions that are in the dataset and outside with histograms
fig, ax = plt.subplots(figsize=(9,6))

sns.histplot(data=dataset_df, x=metric, color='red', stat='probability', binwidth=0.5, ax=ax)
sns.histplot(data=non_dataset_df, x=metric, stat='probability', binwidth=0.5, ax=ax)

ax.legend(labels=['Dataset', 'Non dataset'])

plt.show()
```
