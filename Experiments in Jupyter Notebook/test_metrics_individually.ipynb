```python
# Libraries used
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the files with the data in dataframes for its posterior analysis
dataset_df = pd.read_csv("C:\\Users\\Enrique\\Desktop\\5ºDGMatesInfo\\TFG-Circuit-Complexity\\Experiments\\Jupyter notebooks\\dataset.csv")
non_dataset_df = pd.read_csv("C:\\Users\\Enrique\\Desktop\\5ºDGMatesInfo\\TFG-Circuit-Complexity\\Experiments\\Jupyter notebooks\\non_dataset.csv")
```

```python
X = np.array(list(set(dataset_df["entanglement"])))
Y = np.array([min(dataset_df[dataset_df["entanglement"] == x]["gates"]) for x in X])

fig = plt.figure(figsize=(6,6))
plt.plot(X, Y, 'r')

plt.xlabel("entanglement")
plt.ylabel("gates")
plt.show()
```

```python
min(dataset_df[dataset_df["equanimity_avg"] > 0.5]["gates"])
```

```python
A = non_dataset_df[non_dataset_df["equanimity_subsets"] <= np.percentile(non_dataset_df["equanimity_subsets"], 20)]
B = non_dataset_df[non_dataset_df["equanimity_avg"] >= np.percentile(non_dataset_df["equanimity_avg"], 80)]

set_A = set(A["function"])
set_B = set(B["function"])

len(set_A.intersection(set_B)) / len(set_A)
```

```python
len(non_dataset_df[non_dataset_df["equanimity_avg"] > np.percentile(non_dataset_df["equanimity_avg"], 98)]) / len(non_dataset_df)
```
